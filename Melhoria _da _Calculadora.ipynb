```python
def soma(a, b):
    """Retorna a soma de dois números."""
    return a + b

def subtrai(a, b):
    """Retorna a subtração de dois números."""
    return a - b

def multiplica(a, b):
    """Retorna a multiplicação de dois números."""
    return a * b

def divide(a, b):
    """Retorna a divisão de dois números, com tratamento para divisão por zero."""
    if b == 0:
        return "Erro: divisão por zero"
    return a / b

def mostrar_menu():
    """Exibe o menu de operações."""
    print("+=================+")
    print("| MENU DE OPERAÇÕES")
    print("| 1 - somar        ")
    print("| 2 - subtrair     ")
    print("| 3 - multiplicar  ")
    print("| 4 - dividir      ")
    print("| 0 - sair         ")
    print("+=================+")

def obter_opcao():
    """Obtém e valida a opção do usuário."""
    while True:
        try:
            opcao = int(input("Escolha uma opção: "))
            if opcao in [0, 1, 2, 3, 4]:
                return opcao
            else:
                print("Opção inválida. Tente novamente.")
        except ValueError:
            print("Entrada inválida. Digite um número inteiro.")

def obter_numero(prompt):
    """Obtém e valida um número do usuário."""
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Entrada inválida. Digite um número válido.")

def main():
    while True:
        mostrar_menu()
        opcao = obter_opcao()
        
        if opcao == 0:
            print("Saindo...")
            break

        n1 = obter_numero("Primeiro número: ")
        n2 = obter_numero("Segundo número: ")

        if opcao == 1:
            print("O resultado da soma é:", soma(n1, n2))
        elif opcao == 2:
            print("O resultado da subtração é:", subtrai(n1, n2))
        elif opcao == 3:
            print("O resultado da multiplicação é:", multiplica(n1, n2))
        elif opcao == 4:
            print("O resultado da divisão é:", divide(n1, n2))

if __name__ == "__main__":
    main()
```
